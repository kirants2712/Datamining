**IMPORTING LIBRARIES**


```python
import sys
sys.path.append('/usr/lib/python2.7/dist-packages')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('1.csv')
```

```python
df.info()
```

removing unwanted attributes

```python
del df['Unnamed: 0']
```

```python
del df['Unnamed: 0.1']
```

**HEATMAP OF DATASET BEFORE CLEANING**

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.isnull(),cmap='viridis',cbar=False,yticklabels=False)
```

**FILLING MISSING VALUES WITH MEAN**

```python
df['SE.PRM.TENR'] = df.groupby('Country Name')['SE.PRM.TENR'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['SE.PRM.TENR'] = df.groupby('Region')['SE.PRM.TENR'].transform(lambda x: x.fillna(x.mean()))
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.isnull(),cmap='viridis',cbar=False,yticklabels=False)
```

```python
df['NY.ADJ.NNTY.KD.ZG'] = df.groupby('Region')['NY.ADJ.NNTY.KD.ZG'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['SL.UEM.TOTL.NE.ZS'] = df.groupby('Country Name')['SL.UEM.TOTL.NE.ZS'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['SL.UEM.TOTL.NE.ZS'] = df.groupby('Region')['SL.UEM.TOTL.NE.ZS'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['SE.SEC.ENRL.TC.ZS'] = df.groupby('Income Group')['SE.SEC.ENRL.TC.ZS'].transform(lambda x: x.fillna(x.mean()))
```

```python
del df['Other Groups']
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.isnull(),cmap='viridis',cbar=False,yticklabels=False)
```

```python
df.to_csv('final_cleaned.csv')
```
